```python
# Matches Played in Different Cities
import matplotlib.pyplot as plt
matches_per_city = df['City'].value_counts()
plt.figure(figsize=(14, 7))
plt.bar(matches_per_city.index[:20], matches_per_city.values[:20])  # Show top 20 cities
plt.title('Matches Played in Different Cities (Top 20)')
plt.xlabel('City')
plt.ylabel('Number of Matches')
plt.xticks(matches_per_city.index[:20], rotation=90)
plt.show()
```

```python
# Number of Matches in Different Stages
matches_per_stage = df['Stage'].value_counts()
plt.figure(figsize=(14, 7))
plt.bar(matches_per_stage.index, matches_per_stage.values, color='orange')
plt.title('Number of Matches in Different Stages')
plt.xlabel('Stage')
plt.ylabel('Number of Matches')
plt.xticks(matches_per_stage.index, rotation=45)
plt.show()
```

```python
# Number of Goals Scored by Home Teams vs. Away Teams
goals_data = df[['Home Team Goals', 'Away Team Goals']].sum()
plt.figure(figsize=(10, 5))
plt.bar(goals_data.index, goals_data.values, color=['blue', 'red'])
plt.title('Total Goals Scored by Home Teams vs. Away Teams')
plt.xlabel('Team Type')
plt.ylabel('Total Goals Scored')
plt.show()
```

```python
# Select only the numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix on the numeric DataFrame
correlation_matrix = numeric_df.corr()

print("Correlation coefficients:")
print(correlation_matrix)
```
